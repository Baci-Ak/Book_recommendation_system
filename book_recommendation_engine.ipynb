THE 7 SIMILARITY METRICES FORMULARS:
    
    

1. the minkowski distance similarity formular is, m(x,y) =      $(\sum|x - y|^p)^1/p$


2. the squared euclidean formula, D(x,y) =                       ‚àë(ùë•‚àíùë¶)2


3. spearman's rank correlation, p(x,y) =                         $1 - 6 \sum(d)^2/n(n^2 - 1)$


4. chebyshev =                                                 $max(| x - y|)$


5. hamming distance


6. cosine similarity =                                          $\sum xy / \sqrt \sum x^2  \sqrt \sum y^2$


7. pearson correlation similarity:     $\sum XY - ((\sum X)(\sum Y)/n)/ \sqrt \sum X^2 - ((\sum X)/n)(\sum Y^2 - (\sum Y)^2/n)$

```python
class recommendersystem():
    
    def __init__(self, df):
        self.df = df
        self.user_mapping = {user_id: idx for idx, user_id in enumerate(df['user_id'].unique())}
        self.book_mapping = {book_id: title for book_id, title in zip(df['book_id'], df['booktittle'])}
        self.n_users = len(df['user_id'].unique())
       
        
      
    
    # Implementing Pearson correlation similarity  
    def pearson_similarity(self, user1, user2):
        try:
            common_items = set(self.df[self.df.user_id==user1]['book_id'].dropna()) & set(self.df[self.df.user_id==user2]['book_id'].dropna())
            if not common_items:
                return 0

            user1_ratings = self.df[self.df.user_id==user1]['bookrating'][self.df[self.df.user_id==user1]['book_id'].isin(common_items)].values
            user2_ratings = self.df[self.df.user_id==user2]['bookrating'][self.df[self.df.user_id==user2]['book_id'].isin(common_items)].values
            
            # Check if the length of user1_ratings or user2_ratings is less than 2
            if len(user1_ratings) < 2 or len(user2_ratings) < 2:
                return 0
        
            similarity = scipy.stats.pearsonr(user1_ratings, user2_ratings)[0]
        
            return similarity
        except KeyError:
            print('wrong user ID number, please enter the correct user ID')
        except UnboundLocalError:
            print('a local variable is referenced before assignment, please check your variables and try again')
        #except NameError :
           # print('there is a name in the function that is not defined, please check to fix this')
        except ValueError:
            print("Wrong input of value, please check")
    
    
    #Implementing minkowski similarity
    def minkowski_similarity(self, user1, user2, p = 3):
        try:
            
            
        
            #p = float(input("Enter the value for p in Minkowski similarity: "))
            common_items = set(self.df[self.df.user_id==user1]['book_id'].dropna()) & set(self.df[self.df.user_id==user2]['book_id'].dropna())
            if not common_items:
                return 0
        
            user1_ratings = self.df[self.df.user_id==user1]['bookrating'][self.df[self.df.user_id==user1]['book_id'].isin(common_items)].values
            user2_ratings = self.df[self.df.user_id==user2]['bookrating'][self.df[self.df.user_id==user2]['book_id'].isin(common_items)].values
        
            distance = minkowski(user1_ratings, user2_ratings, p)
        
            if distance == 0:
                return 1 / (1e-9)  # Avoid division by zero
            return round(1 / distance, 2)
        
        except KeyError:
            print('wrong user ID number, please enter the correct user ID')
        except UnboundLocalError:
            print('a local variable is referenced before assignment, please check your variables and try again')
        #except NameError :
            #print('there is a name in the function that is not defined, please check to fix this')
            
        except UnboundLocalError:
            print('a local variable is referenced before assignment, please check your variables and try again')
    
    
    
    def cosine_similarity(self, user1, user2):
        try:
            common_items = set(self.df[self.df.user_id==user1]['book_id'].dropna()) & set(self.df[self.df.user_id==user2]['book_id'].dropna())
            if not common_items:
                return 0
            
            user1_ratings = self.df[self.df.user_id==user1]['bookrating'][self.df[self.df.user_id==user1]['book_id'].isin(common_items)].values.reshape(1, -1)
            user2_ratings = self.df[self.df.user_id==user2]['bookrating'][self.df[self.df.user_id==user2]['book_id'].isin(common_items)].values.reshape(1, -1)
            
            similarity_mat = round(cosine_similarity(user1_ratings, user2_ratings)[0][0], 2)
            return similarity_mat
        except KeyError:
            print('wrong user ID number, please enter the correct user ID')
        except UnboundLocalError:
            print('a local variable is referenced before assignment, please check your variables and try again')
        #except NameError :
            #print('there is a name in the function that is not defined, please check to fix this')
        except ValueError:
            print("Wrong input of value, please check")
            
    
    
    def spearman_correlation_similarity(self, user1, user2):
        try:
            common_items = set(self.df[self.df.user_id==user1]['book_id'].dropna()) & set(self.df[self.df.user_id==user2]['book_id'].dropna())
            if not common_items:
                return 0
            
            user1_ratings = self.df[self.df.user_id==user1]['bookrating'][self.df[self.df.user_id==user1]['book_id'].isin(common_items)].values
            user2_ratings = self.df[self.df.user_id==user2]['bookrating'][self.df[self.df.user_id==user2]['book_id'].isin(common_items)].values
            
            # Check if the length of user1_ratings or user2_ratings is less than 2
            if len(user1_ratings) < 2 or len(user2_ratings) < 2:
                return 0
            spearman_similarity = round(scipy.stats.spearmanr(user1_ratings, user2_ratings)[0], 3)
   
            return spearman_similarity
        
        except KeyError:
            print('wrong user ID number, please enter the correct user ID')
        except UnboundLocalError:
            print('a local variable is referenced before assignment, please check your variables and try again')
        #except NameError :
           # print('there is a name in the function that is not defined, please check to fix this')
    
            
     # n_similar_user method       
    def n_similar_users(self, target_user, n=5, metric='cosine'):
        try:
            
            user_similarities = {}
            for user in self.df['user_id'].unique():
                if user != target_user:
                
                    if metric == 'pearson':
                        similarity = self.pearson_similarity(target_user, user)
                    
                    elif metric == 'minkowski':

                        similarity = self.minkowski_similarity(target_user, user)
                    
                    elif metric == 'spearman':
                        similarity = self.spearman_correlation_similarity(target_user, user)
                
                    else:
                        similarity = self.cosine_similarity(target_user, user)
                    
                    user_similarities[user] = similarity

                   # Sort users based on similarity and get top n similar users
            sorted_similar_users = sorted(user_similarities.items(), key=lambda x: x[1], reverse=True)
        
                     # to return similar users with the similarity score, use below commented lind of code
                      #n_similar_users = [(user, similarity) for user, similarity in sorted_similar_users[:n]]
            n_similar_users = [user for user, similarity in sorted_similar_users[:n]]
            print('\n')

            return n_similar_users
        
        except KeyError:
            print('wrong user ID number, please enter the correct user ID')
        except UnboundLocalError:
            print('a local variable is referenced before assignment, please check your variables and try again')
        except NameError :
            print('there is a name in the function that is not defined, please check to fix this')
            
        except UnboundLocalError:
            print('a local variable is referenced before assignment, please check your variables and try again')
    
    
 
    #recommendation method
    
    def make_recommendation(self, target_user, n=5, metric='cosine'):
        try:
            similar_users = self.n_similar_users(target_user, n=n, metric=metric)

                  # Find books that similar users have read and target user hasn't read
            recommended_books = {}
            for user in similar_users:
                for booktittle in self.df[(self.df['user_id'] == user) & (self.df['bookrating'] >=3)]['booktittle'].values:
                    if not self.df[(self.df['user_id'] == target_user) & (self.df['booktittle'] == booktittle)]['booktittle'].values:
                        if booktittle in recommended_books:
                            recommended_books[booktittle] += 1
                        else:
                            recommended_books[booktittle] = 1

                     # Sort recommended books by count
            sorted_recommended_books = sorted(recommended_books.items(), key=lambda x: x[1], reverse=True)
            recommendations = pd.DataFrame(list(dict(sorted_recommended_books[:n]).items()), columns=['Recommendations', 'recommendationCount'])

            return recommendations
        
        except KeyError:
            print('wrong user ID number, please enter the correct user ID')
        except UnboundLocalError:
            print('a local variable is referenced before assignment, please check your variables and try again')
        except NameError :
            print('there is a name in the function that is not defined, please check to fix this')
            
        except UnboundLocalError:
            print('a local variable is referenced before assignment, please check your variables and try again')
    

    # metric evaluation method
    
    def evaluate_similarity_metrics_cor(self, target_user, n=5, similarity_metrics=['cosine']):
        recommendation_accuracies = {}

        for metric in similarity_metrics:
            similar_users = self.n_similar_users(target_user, n=n, metric=metric)
            recommendations = self.make_recommendation(target_user, n=n, metric=metric)

            if recommendations.empty:
                print(f"No recommendations found for metric {metric}. Skipping...")
                recommendation_accuracies[metric] = 0.0
                continue

                 # Calculate correct recommendations for all similar users
            correct_recommendations = 0
            for user in similar_users:
                similar_user_books = set(self.df[(self.df['user_id'] == user) & (self.df['bookrating'] >=3)]['booktittle'].values)
                if recommendations['Recommendations'].isin(similar_user_books).any():
                    correct_recommendations += 1

                     # Calculate the recommendation accuracy
            accuracy = correct_recommendations / n
            recommendation_accuracies[metric] = accuracy
            recommendation_accuracy = pd.DataFrame(recommendation_accuracies.items(), columns=['metric', 'Accuracy'])

        return recommendation_accuracy
    
    
    
    
    
    
 
```
